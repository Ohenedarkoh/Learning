```python
import pandas as pd 
```

```python
dict1 = {'Name': ['Priyang Bhatt', 'Rashmin Panchal', 'Anil Rana', 'Rahul Patel'],
        'City': [' New York ', ' Los Angeles ', ' Chicago', ' Houston'],
        'State': ['NY', 'CA', 'IL', 'TX']}
data = pd.DataFrame(dict1)
data
```

```python
#str.lower()


data['Name_lower']=data['Name'].str.lower()
data
```

```python
data['Name_upper']=data['Name'].str.upper()
data
```

```python
##str.lenght


data['name_len']=data['Name'].str.len()
data
```

```python
##str.strip

data['City'][0]
```

```python
data['City_strip']=data['City'].str.strip()
data
```

```python
data['right_strip']=data['City'].str.rstrip()
data
```

```python
data['left_strip']=data['City'].str.lstrip()
data
```

```python
#str.split()- allows to split a string into a list of strings

data
```

```python
data['Name']
```

```python
data[['First_name', 'Last_name']]=data['Name'].str.split(' ', expand=True)
data
```

```python
#str.contains()
data[data['Name'].str.contains('Bhatt')]
```

```python
#str.replace()-used to replace one or more sub strings in a str with another str
```

```python
data
```

```python
data['state_replaced']=data['State'].str.replace('NY','New York').str.replace('CA','California')
data
```

```python
#str.startswith()- used to check is a str starts with a specific sub string or not
data['starts_with'] =data['First_name'].str.startswith('P')
data
```

```python
data['ends_with_l']=data['Last_name'].str.endswith('l')
data
```

```python
#str.cat()-concatenate
data
```

```python
data['full_name']=data['First_name'].str.cat(data['Last_name'],sep=' ')
data
```
