```python
#merging data frames


import pandas as pd
```

```python
#reading csv file
#saving table 1 in a variable 

df1 = pd.read_csv(r"C:\Users\ACER\Desktop\Learning\data\personal\python\LOTR.csv")
df1
```

```python
#reading csv file
#saving table 2 in a variable 



df2 = pd.read_csv(r"C:\Users\ACER\Desktop\Learning\data\personal\python\LOTR 2.csv")
df2  
```

```python
#inner merging table one and 2. Inner merge only return a table that has the exact columns in the two tables


df1.merge(df2, how='inner', on =['FellowshipID','FirstName'])
```

```python
#outer merge merges all columns in the two tables


df1.merge(df2 , how = 'outer')
```

```python
#left join add any columns that adds all columns from the right table and those that are exceptions too


df1.merge(df2 , how = 'left')
```

```python
#merging all columns and used to compare values



df1.merge(df2 , how = 'cross')
```

```python
#outer joining and adding suffixes to  a column header 



df1.join(df2 , on = 'FellowshipID' , how='outer' , lsuffix = '_left' , rsuffix = '_Right')
```

```python

#setting index and adding suffixes



df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID') , lsuffix = '_left' , rsuffix = '_Right'  , how = 'outer')
df4
```

```python
#concatinating tables on outer join


pd.concat([df1,df2] , join = 'outer' , axis = 1)
```
