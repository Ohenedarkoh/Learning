```python
#data visualisations
#importing all visualisation libraries




import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the csv file and setting index (giving it a variable)


df = pd.read_csv(r"C:\Users\ACER\Desktop\Learning\data\personal\python\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
#choosing plot/chart style


print(plt.style.available)
plt.style.use('fivethirtyeight')
df
```

```python

#specifying plot/chart style and size


df.plot(kind = 'line', figsize =(10,5))
```

```python
#separating charts into subplots


df.plot(kind = 'line' ,subplots = True)
```

```python
#after specifying the type of chart, give it a title, an x label and a y label name



df.plot(kind = 'line' , title = 'Ice Cream Ratings' , xlabel ='Daily Ratings' , ylabel = 'Scores' )
```

```python
#bar chart

df.plot(kind = 'bar')
```

```python
#stacked bar chart

df.plot(kind = 'bar' , stacked =True)
```

```python
#stacked bar chart but for only one intereted variable



df['Flavor Rating'].plot(kind = 'bar' , stacked =True)
```

```python
#horizontal stacked bar chart


df.plot.barh(stacked =True)
```

```python
#scatter plot with specified variables of interest, size and colour


df.plot.scatter(x = 'Texture Rating' , y= 'Overall Rating' ,  s=500 , c ='Yellow')
```

```python
#histogram and bins



df.plot.hist( bins = 10)
```

```python
#boxplot


df.plot.box()
```

```python
#area plot



df.plot.area( figsize =(10 ,5))
```

```python
#pie chart of a varibale and size



df.plot.pie(y = 'Flavor Rating' , figsize= (10,6))
```
