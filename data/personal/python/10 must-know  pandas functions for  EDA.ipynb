```python
import pandas as pd
```

```python
data = {'age': [25, 22, 18, 30, 45, 50, 35, 20, 55, 40],
          'gender': ['M', 'F', 'F', 'M', 'M', 'F', 'M', 'F', 'M', 'M'],
            'score': [90, 80, 75, 95, 70, 85, 75, 90, 95, 85]}
df = pd.DataFrame(data)
df
```

```python
#VALUE_COUNTS FUNCTION

df['gender'].value_counts()
```

```python
df['age'].value_counts()
```

```python
#WHERE FUNCTION


df[df.where(df['age']>30,other =0).all(1)]

```

```python
f1 = df['gender'] =='F'
f2 = df['score']>30
```

```python
df[df.where(f1 & f2 , other = 0).all(1)]
```

```python
#ISIN 


df[df['age'].isin([25,35])]
```

```python
df[df[['age','gender']].isin({'age':[25,35],'gender':['M']}).all(1)]
```

```python
#CUT AND QCUT (cut-used to divide continuous variable into discrete bins) 
df['score_bins']=pd.cut(df['score'],bins=[60,70,80,90,100])
df
```

```python

bin_names=['low','good','very good','high','very high']
df['score_bins']=pd.cut(df['score'],bins=[50,60,70,80,90,100],labels=bin_names)
df
```

```python
df
```

```python
#qcut is used to divide a set of continuous variables into  sets of discrete bins 



age_bins=pd.qcut(df['age'],4)
age_bins
```

```python

age_bins=pd.qcut(df['age'],4)
age_bins
```
